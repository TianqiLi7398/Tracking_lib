# Sequential JPDA

After implementing the JPDA in single sensor, now we are interested in multisensor system. Sequential JPDA is the centralized version of JPDA in multisensor system.

## 1. Generate observation data
Now first let's get some inputs. We simulate some random points plus a sin wave trajectory in 2D Cartesian space.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import random
import scipy.linalg
from numpy.linalg import inv
import copy
from scipy.stats import multivariate_normal
from scipy.stats.distributions import chi2
from numpy import linalg as LA
from matplotlib.patches import Ellipse
from matplotlib.patches import Rectangle
import math
```

Here we will have 3 tracking for 3 objects doing:
    1. sin wave movement
    2. linear movement
    3. Parabola

```python
input_var = 0.1
dt = 0.1
time_range = np.linspace(dt, 20, 200)

obs_list = []
obs_list_1 = []
obs_list_2 = []
obs_list_3 = []
obs_list_4 = []
random.seed(1)
traj1 = {"curve": 'sin','time_interval': [0, 20], "points": []}
traj2 = {"curve": 'line','time_interval': [0, 11], "points": []}
traj3 = {"curve": 'circle','time_interval': [8, 20], "points": []}
traj = [traj1, traj2, traj3]
noise = []
fov1 = ['rectangle', [0,0], [15, 10]] # [shape, leftbot pt, rightup pt]
fov2 = ['rectangle', [15,0], [30, 10]]
fov3 = ['rectangle', [0,-10], [15, 0]]
fov4 = ['rectangle', [15,-10], [30, 0]]
fov_list = [fov1, fov2, fov3, fov4]
for traj_ in traj:
    [t0, t1] = traj_["time_interval"]
    if np.isclose(t0, 0):
        t0 = dt
    traj_["time_range"] = np.linspace(t0, t1, int(np.floor(t1/dt - t0/dt)) + 1)

def inside_time_interval(t, interval):
    if t >= interval[0]:
        if t <= interval[1]:
            return True
    return False

for t in time_range:
    # the sin wave signal
    z_k = []
    noise_k = []
    z_k1 = []
    z_k2 = []
    z_k3 = []
    z_k4 = []

    if inside_time_interval(t, traj[0]["time_interval"]):

        point = [t * 1.5 + 0.1 * np.random.normal(0, input_var), 3 + 3 * np.sin(t * 1.5) -2.5 + 0.1 * np.random.normal(0, input_var)]
        traj[0]["points"].append(point)
        if (point[0] >= fov1[1][0] and point[0] <= fov1[2][0] and point[1] >= fov1[1][1] and point[1] <= fov1[2][1]):
            z_k1.append(point)
        elif (point[0] >= fov2[1][0] and point[0] <= fov2[2][0] and point[1] >= fov2[1][1] and point[1] <= fov2[2][1]):
            z_k2.append(point)
        elif (point[0] >= fov3[1][0] and point[0] <= fov3[2][0] and point[1] >= fov3[1][1] and point[1] <= fov3[2][1]):
            z_k3.append(point)
        elif (point[0] >= fov4[1][0] and point[0] <= fov4[2][0] and point[1] >= fov4[1][1] and point[1] <= fov4[2][1]):
            z_k4.append(point)
        z_k.append(point)
    
    if inside_time_interval(t, traj[1]["time_interval"]):
        point = [3 * t +  np.random.normal(0, input_var),  t - 5 +  np.random.normal(0, input_var)]
        traj[1]["points"].append(point)
        if (point[0] >= fov1[1][0] and point[0] <= fov1[2][0] and point[1] >= fov1[1][1] and point[1] <= fov1[2][1]):
            z_k1.append(point)
        elif (point[0] >= fov2[1][0] and point[0] <= fov2[2][0] and point[1] >= fov2[1][1] and point[1] <= fov2[2][1]):
            z_k2.append(point)
        elif (point[0] >= fov3[1][0] and point[0] <= fov3[2][0] and point[1] >= fov3[1][1] and point[1] <= fov3[2][1]):
            z_k3.append(point)
        elif (point[0] >= fov4[1][0] and point[0] <= fov4[2][0] and point[1] >= fov4[1][1] and point[1] <= fov4[2][1]):
            z_k4.append(point)
        z_k.append(point)
    
    if inside_time_interval(t, traj[2]["time_interval"]):
        point = [15 + 4 * np.cos(t) + 0.1 * np.random.normal(0, input_var),  4 * np.sin(t) + 0.1 * np.random.normal(0, input_var)]
        traj[2]["points"].append(point)
        if (point[0] >= fov1[1][0] and point[0] <= fov1[2][0] and point[1] >= fov1[1][1] and point[1] <= fov1[2][1]):
            z_k1.append(point)
        elif (point[0] >= fov2[1][0] and point[0] <= fov2[2][0] and point[1] >= fov2[1][1] and point[1] <= fov2[2][1]):
            z_k2.append(point)
        elif (point[0] >= fov3[1][0] and point[0] <= fov3[2][0] and point[1] >= fov3[1][1] and point[1] <= fov3[2][1]):
            z_k3.append(point)
        elif (point[0] >= fov4[1][0] and point[0] <= fov4[2][0] and point[1] >= fov4[1][1] and point[1] <= fov4[2][1]):
            z_k4.append(point)
        z_k.append(point)

    
    for i in range(random.randint(5, 10)):
        point = [30* random.random(), 20 * random.random() - 10]
        if (point[0] >= fov1[1][0] and point[0] <= fov1[2][0] and point[1] >= fov1[1][1] and point[1] <= fov1[2][1]):
            z_k1.append(point)
        elif (point[0] >= fov2[1][0] and point[0] <= fov2[2][0] and point[1] >= fov2[1][1] and point[1] <= fov2[2][1]):
            z_k2.append(point)
        elif (point[0] >= fov3[1][0] and point[0] <= fov3[2][0] and point[1] >= fov3[1][1] and point[1] <= fov3[2][1]):
            z_k3.append(point)
        elif (point[0] >= fov4[1][0] and point[0] <= fov4[2][0] and point[1] >= fov4[1][1] and point[1] <= fov4[2][1]):
            z_k4.append(point)
        noise_k.append(point)

    noise.append(noise_k)
    z_k += noise_k
        
    obs_list.append(z_k)
    
    obs_list_1.append(z_k1)
    obs_list_2.append(z_k2)
    obs_list_3.append(z_k3)
    obs_list_4.append(z_k4)
multi_sensor_obs = [obs_list_1, obs_list_2, obs_list_3, obs_list_4]
```

```python
def make_ellipse(mean, cov, gating_size):
    """Support function for scatter_ellipse."""
    

    v, w = np.linalg.eigh(cov)
    u = w[0] / np.linalg.norm(w[0])
    angle = np.arctan(u[0, 1]/u[0,0])
    angle = 180 * angle / np.pi # convert to degrees
    v = 2 * np.sqrt(v * gating_size) #get size corresponding to level
    return Ellipse(mean[:2], v[0], v[1], 180 + angle, facecolor='none',
                  edgecolor='green',
                  #ls='dashed',  #for debugging
                  lw=1.5)
```

```python
# set up figure and animation
fig = plt.figure()
# fig.subplots_adjust(left=0, right=1, bottom=0, top=1)
ax = fig.add_subplot(111, aspect='equal', autoscale_on=False,
                    xlim=(-2, 32), ylim=(-12, 12))

fov_pt1, = ax.plot([], [], 'bo', ms=2)
fov_pt2, = ax.plot([], [], 'yo', ms=2)
fov_pt3, = ax.plot([], [], 'ro', ms=2)
fov_pt4, = ax.plot([], [], 'go', ms=2)
fov_pt_list = [fov_pt1, fov_pt2, fov_pt3, fov_pt4]
# obs_list_anim = np.array(obs_list)
line1, = ax.plot([], [], '--', lw=1, color = 'b')
line2, = ax.plot([], [], '--', lw=1, color = 'y')
line3, = ax.plot([], [], '--', lw=1, color = 'r')

real_point1_est, = ax.plot([], [], 'b*', ms=2)
real_point2_est, = ax.plot([], [], 'y*', ms=2)
real_point3_est, = ax.plot([], [], 'r*', ms=2)

# sensor_fov_draw_circle = plt.Circle((sensor_pos[0], sensor_pos[1]), shape[1], fill = False)


rand_point, = ax.plot([], [], 'go', ms=6)
obs_list_anim = np.array(obs_list)

time_text = ax.text(0.05, 0.9, '', transform=ax.transAxes)


def init():
    """initialize animation"""
    
    fov_pt1.set_data([], [])
    fov_pt2.set_data([], [])
    fov_pt3.set_data([], [])
    fov_pt4.set_data([], [])

    line1.set_data([], [])
    line2.set_data([], [])
    line3.set_data([], [])
    
    for fov in fov_list:
        ax.add_artist(Rectangle((fov[1][0], fov[1][1]), 
                        fov[2][0] - fov[1][0], fov[2][1] - fov[1][1], 
                        fc ='none',  
                        ec ='g', 
                        lw = 1,
                        linestyle = '-'))
    
    #     ax.add_artist(sensor_fov_draw_circle)
    return fov_pt1, fov_pt2, fov_pt3, fov_pt4, line1, line2, line3

def animate(i):
    """perform animation step"""
    global dt, ax, fig
    
    dt = 0.1
    
#     x, y = [], []
#     for point in jpda_est[i]:
#         x.append(point[0, 0])
#         y.append(point[1, 0])
#     real_point1_est.set_data(x, y)


    # update pieces of the animation
    if inside_time_interval(i*dt + dt, traj[0]['time_interval']):
        index = int((i*dt - traj[0]['time_interval'][0])/dt)
        real_point1.set_data(traj[0]['points'][index][0], traj[0]['points'][index][1])
        line1.set_data(1.5*traj[0]["time_range"], 3 + 3 * np.sin(1.5*traj[0]["time_range"]) -2.5)
    
    if inside_time_interval(i*dt + dt, traj[1]['time_interval']):
        index = int((i*dt - traj[1]['time_interval'][0])/dt)
        
        real_point2.set_data(traj[1]['points'][index][0], traj[1]['points'][index][1])
        line2.set_data(3*traj[1]["time_range"], traj[1]["time_range"] - 5)
    
    if inside_time_interval(i*dt + dt, traj[2]['time_interval']):
        index = i - int(traj[2]['time_interval'][0]*10) + 1
        
        real_point3.set_data(traj[2]['points'][index][0], traj[2]['points'][index][1])
        line3.set_data(15 + 4 *np.cos(time_range), 4 * np.sin(time_range))
    
    
#     x , y = [], []
#     for point in noise[i]:
        
#         x.append(point[0])
#         y.append(point[1])
#     rand_point.set_data(x, y)
    for j in range(len(fov_list)):
        x , y = [], []
        for point in multi_sensor_obs[j][i]:
        
            x.append(point[0])
            y.append(point[1])
        fov_pt_list[j].set_data(x, y)
#     #patches = []
#     for obj in ax.findobj(match = Ellipse):
#         obj.remove()
#     for gates in ellips_inputs[i]:
#         mu = [gates[1][0,0], gates[1][1,0]]
#         S = gates[0]
#         e = make_ellipse(mu, S, sensor1.gating_size)
#         ax.add_artist(e)
    
#     gates = Ellipse(sensor_pos, 2*shape[1][0], 2*shape[1][0], facecolor='none',
#                 edgecolor='black',
#                 #ls='dashed',  #for debugging
#                 lw=1.5)
    
#     ax.add_artist(gates)
    
    return fov_pt1, fov_pt2, fov_pt3, fov_pt4, line1, line2, line3#, real_point1_est, real_point2_est, real_point3_est#, time_text, patches

ani = animation.FuncAnimation(fig, animate, frames=len(obs_list),
                            interval=10, blit=True, init_func=init, repeat = False)
#plt.show()

ani.save('video/seq_jpda.mp4', fps=20)
```
